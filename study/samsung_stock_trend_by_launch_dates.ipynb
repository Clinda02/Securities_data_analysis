## 데이터 설정 :
1. 삼성전자 주가 데이터 불러오기
2. 갤럭시 출시일 설정

```python
# 삼성전자 주가 데이터 설정
"""
yf.download 모듈 사용하여
삼성전자 주가(005930.KS) 데이터 3년치(2022-01-01 ~ 2024-01-01) 다운로드 하기
"""
df = yf.download('005930.KS', start='2021-01-01', end='2024-01-01')
df.index = df.index.tz_convert('UTC')

# 갤럭시 출시일 설정
launch_galaxy_dates = {
    "갤럭시 S21": "2021-01-29",
    "갤럭시 Z Flip/Fold 3": "2021-08-27",
    "갤럭시 S22": "2022-02-25",
    "갤럭시 Z Flip/Fold 4": "2022-08-26",
    "갤럭시 S23": "2023-02-17",
    "갤럭시 Z Flip/Fold 5": "2023-07-26",
}
launch_dates = {name: pd.to_datetime(date).tz_localize('UTC') for name, date in launch_galaxy_dates.items()}
```

## 데이터 시각화(1) :
1. 삼성전자 주가 line 그래프 그리기
2. 그래프 내, 출시일 표시
3. 그래프 제목 설정 : "갤럭시 출시에 따른 삼성전자 주가 변동 그래프"
4. x축 레이블 : "날짜"
5. y축 레이블 : "주가 (KRW)"
6. 범례(legend) 추가
7. 그리드(grid) 추가

## 예시 결과 :
완성본 데이터.png

```python
# 코드 작성

df.index = pd.to_datetime(df.index)

plt.figure(figsize=(10, 6))

# 삼성전자 주가 그래프
plt.plot(df.index, df['Close'], label = "삼성전자 주가", color = 'blue')

# 삼성전자 갤럭시 출시일 표시
for name, date in launch_galaxy_dates.items():
  date = pd.to_datetime(date)
  plt.axvline(x=date, label = name , color = 'red', linestyle = '--')

plt.title("갤럭시 출시에 따른 삼성전자 주가 변동 그래프")
plt.xlabel("날짜")
plt.ylabel("주가 (KRW)")
plt.legend()
plt.grid()

plt.show()
```

## 데이터 시각화(2) :
1. 각 출시일에 대한 서브플롯 생성
2. 각 출시일 날짜 기준 (-3일, +7일) 데이터 생성
    (plt.axvline() 함수 사용)
3. 위 데이터 기준 서브플롯 그래프 그리기
4. 각 그래프 제목 및 레이블, 범례, 그리드 설정

## 예시 결과 :
image.png

```python
# 코드 작성
num_launches = len(launch_dates)
fig, axes = plt. subplots(num_launches, 1, figsize=(14, num_launches *4 ))

for ax, (name, date) in zip(axes, launch_dates.items()):
  start_date =  date - pd.Timedelta(days =3)
  end_date = date + pd.Timedelta(days =7)

  sub_df = df.loc[start_date:end_date]

  ax.plot(sub_df.index, sub_df['Close'], label="삼성전자 주가", color = 'blue')
  ax.axvline(x=date, label = name , color = 'red', linestyle = '--')

  ax.set_title(f"{name}출시일 전후 주가 변동")
  ax.set_xlabel("Date")
  ax.set_ylabel("Price")
  ax.legend()
  ax.grid()

plt.show()

```

## 데이터 시각화(3) :
1. 경쟁사 아이폰 출시일도 표시
2. 환율 데이터도 추가하여 y축에 표시 (twinx() 함수 사용)

## 예시 결과 :
image.png

```python
galaxy_launch = {
    "갤럭시 S21": "2021-01-29",
    "갤럭시 Z Flip/Fold 3": "2021-08-27",
    "갤럭시 S22": "2022-02-25",
    "갤럭시 Z Flip/Fold 4": "2022-08-26",
    "갤럭시 S23": "2023-02-17",
    "갤럭시 Z Flip/Fold 5": "2023-07-26",
}
iphone_launch = {
    "아이폰 13": "2021-09-24",
    "아이폰 14": "2022-09-16",
    "아이폰 15": "2023-09-22",
}

galaxy_launch_dates = {name: pd.to_datetime(date).tz_localize('UTC') for name, date in galaxy_launch.items()}
iphone_launch_dates = {name: pd.to_datetime(date).tz_localize('UTC') for name, date in iphone_launch.items()}

exchange_rate = yf.download('KRW=X', start='2021-01-01', end='2024-01-01')
exchange_rate.index = exchange_rate.index.tz_convert('UTC')
```

```python
# 코드 작성

fig, ax1 = plt.subplots(figsize=(14,7))

# 삼성전자 주가 그래프
ax1.plot(df.index, df['Close'], label = "삼성전자 주가", color = 'navy')
ax.set_xlabel("날짜")
ax.set_ylabel("주가 (KRW)")

# 삼성전자 갤럭시 출시일 표시
for name, date in launch_galaxy_dates.items():
  date = pd.to_datetime(date)
  plt.axvline(x=date, label = name , color = 'red', linestyle = '--')

# 아이폰 출시일
for name, date in iphone_launch_dates.items():
  date = pd.to_datetime(date)
  plt.axvline(x=date, label = name , color = 'royalblue', linestyle = ':')

 # x축(날짜)을 공유하는 두 번째 y축(환율) 생성
ax2 = ax1.twinx()
ax2.plot(exchange_rate.index, exchange_rate['Close'], label='환율', color='orange')

plt.title("삼성전자 주가 변동 및 갤럭시/아이폰 출시일과 환율")


# 범례 표기
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()

handles = handles1 + handles2
labels = labels1 + labels2
plt.legend(handles, labels)

plt.grid()
plt.show()
```
